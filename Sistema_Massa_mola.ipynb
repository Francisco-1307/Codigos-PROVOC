```python
import numpy as np
import matplotlib.pyplot as plt
import time
from matplotlib import animation
from IPython.display import HTML
```

```python
# Variáveis do problema
k = 1. # Constante da mola: unidade N/m 
m = 1. # Massa do corpo   : unidade Kg

T  = 20.       # Tempo total da simulação: unidade s
dt = 0.001     # Intervalo de tempo: unidade s
N  = int(T/dt) # Número de 'frames'

# Condições iniciais
x0 = 1.
v0 = 0.
```

```python
# Arrays de posição e velocidade
x = np.zeros(N+1)
x[0] = x0          # Estabelece condição inicial x0  
v = np.zeros(N+1)
v[0] = v0          # Estabelece condição inicial v0 
```

```python
# Exemplo
np.array([n**2 for n in range(10+1)]) # List comprehension 
```

```python
# Array com os instantes de tempo: de 0 até T com intervalo dt
instantes = np.array([n*dt for n in range(N+1)])
instantes
```

```python
# Intregação da 2a Lei de Newton
for n in range(N): # Para n variando de 0 até N-1
  a = -(k/m) * x[n] # Aceleração no instante n: a = F/m
  v[n+1] = v[n] + a * dt 
  x[n+1] = x[n] + v[n] * dt
```

```python
# Vamos fazer gráficos da velocidade e posição 
plt.figure(figsize = (8,8))            # Define o tamanho da figura abaixo
plt.plot(instantes, x, color = 'blue') # Faz um gráfico t vs x(t)
plt.plot(instantes, v, color = 'red' ) # Faz um gráfico t vs v(t)
plt.xlabel("t")                        # Define o nome do eixo horizontal. Note que o nome é uma string.
plt.grid()                             # Coloca essa malha ao fundo
```

Solução exata: $x(t)  = x_0 cos(\omega t)$ com $\omega = \sqrt{\frac{k}{m}}$

```python
# Comparar com a solução exata
omega = np.sqrt(k/m)
def x_exato(t):
  return x0*np.cos(omega*t)

plt.figure(figsize = (8,8))                              # Define o tamanho da figura abaixo
plt.plot(instantes, x, color = 'blue')                   # Faz um gráfico t vs x(t)
plt.plot(instantes, x_exato(instantes), color = 'black') # Faz um gráfico t vs x_exato(t)
plt.xlabel("t")                                          # Define o nome do eixo horizontal. Note que o nome é uma string.
plt.grid()                                               # Coloca essa malha ao fundo
```
